## Finding the Binding Energy per nucleon for any A and Z

```python
from numpy import sqrt
A = float(input("Mass Number: "))
Z = float(input("Atomic Number: "))
a1 = 15.67
a2 = 17.23
a3 = .75
a4 = 93.2

if A % 2 != 0:
    a5 = 0
elif A % 2 == 0 and Z % 2 == 0:
    a5 = 12
else:
    a5 = -12
#This is the solution to Part A
B = a1 * A - a2 * A ** (2/3) - a3 * Z ** 2 / (A ** (1/3)) - a4 * (A -2 * Z) ** 2 / A + a5 / sqrt(A)
print(f"Binding energy: {B:.3} MeV")

#this is the modification specified for part B
bpa= B / A
print(f"Binding energy per nucleon: {bpa:.3} MeV per nucleon")
```

## Here we will find the value of $A$ with the highest binding energy for any given $Z$

```python
#this is all of the constants and the equation for B as found in the book. I just typed it in here
from numpy import sqrt
Z = int(input("Atomic Number: "))
a1 = 15.67
a2 = 17.23
a3 = .75
a4 = 93.2

Blist = []
Alist = []

for A in range(Z, 3 * Z):    #For loop to look for all the different values of B
    if A % 2 != 0:
        a5 = 0
    elif A % 2 == 0 and Z % 2 == 0:
        a5 = 12
    else:
        a5 = -12
    B = a1 * A - a2 * A ** (2/3) - a3 * Z ** 2 / (A ** (1/3)) - a4 * (A -2 * Z) ** 2 / A + a5 / sqrt(A)
    Blist.append(B/A)      #This stores all the possible values of B in a list so that my next few lines
    Alist.append(A)      #Can analyze them to get me highest B value, and tell which A gives me that value.

bind = max(Blist)
Abest = Alist[Blist.index(bind)]
print(f"Atomic number with highest binding energy per nucleon: {Abest}")
print(f"Binding energy per nucleon: {bind:.3}")
```

## We now modify code so that it runs through all possible values of $Z$ between $1$ and $100$ and each posible value of $A$ for those $Z$'s in order to determine at which value of $Z$ the maximum binding energy per nucleon will occur.

## Here we have simple exercise in finding the coefficients of Pascal's Triangle using lists.

```python
from math import factorial

def binomial(n,d):
    if d ==0:      #the 1st digit in the row is a 1
        return 1
    elif d > n:     #if the digit is the last one on its row, then it is automatically a 1
        return 1 
    else:
        coefficient = factorial(n) / (factorial(d) * factorial(n - d))
    return int(coefficient)                #Here its converted to an integer,simply for the format at the end

r = 1                     #The function defined above will find Pascals triangle, where r = the row number
d = 0                      # and d = digit (Not the digit value itself, but its place in the triangle)
row = []
for r in range(1, 21):               #This for loop goes through each of the 20 rows
    for k in range(0, r + 1):          #This for loop is for the digit values within each of the rows
        row.append(binomial(r, d))      
        d += 1                          
        k += 1
    print("Line " + str(r) + ": " + str(row))
    r += 1
    d = 0
    row = []
```

## We can even use this method to calculate the probabillity of a certain outcome happening, which can be useful in quantum mechanics.

```python
from math import factorial

def binomial(n,k):   #further explanations of this function are in Part B
    if k == 0:
        return 1
    elif k > n:
        return 1
    else:
        coefficient = factorial(n) / (factorial(k) * factorial(n - k))
    return coefficient

n = 100 #number of times the coin was tossed
k = 60 #number of time the coin came up heads

#the probability of this happening is calculated as follows:
prob1 = binomial(n, k) / (2 ** n) * 100
print(f"The probabillity of a coin tossed 100 times coming up heads 60 times is: %{prob1:.2}") 

#The probability of a coin landing heads 60 or more times is the same as the probability of the
#coin landing heads 60 times, + probability of 61 times, etc. We can solve this with a for loop
prob2 = 0
for k in range(60, 100):
    prob2 += binomial(n, k) / (2 ** n) * 100
    k += 1
print(f"The probabillity of a coin tossed 100 times coming up heads 60 times or more is: %{prob2:.2}") 
```
